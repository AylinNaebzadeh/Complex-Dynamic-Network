```python
import networkx as nx
import EoN
import matplotlib.pyplot as plt
```

The SIS (Susceptible-Infected-Susceptible) and SIR (Susceptible-Infected-Recovered) models are two types of compartmental models used in epidemiology to simulate the spread of infectious diseases ².

The main difference between the two models is the assumption about immunity. In the SIR model, individuals who recover from the disease are assumed to have lifelong immunity and enter a "Recovered" compartment where they can no longer be infected or spread the disease ². This model is often used for diseases like measles, mumps, and rubella where recovery confers lasting resistance ².

In contrast, the SIS model assumes that individuals who recover from the disease do not have immunity and can become infected again ¹. This model is often used for diseases like gonorrhea where individuals can get repeatedly infected ¹.

```python
G = nx.fast_gnp_random_graph(1000, 0.002)
t, S, I = EoN.basic_discrete_SIS(G, 0.6, tmax = 20)
plt.plot(t,S)
```

The `EoN.basic_discrete_SIR` function performs a simple discrete SIR (Susceptible-Infected-Recovered) simulation on a given network `G` with a constant transmission probability `p` ¹. The function returns four numpy arrays: `t`, `S`, `I`, and `R`. These arrays give all the times observed and the number of individuals in each state (Susceptible, Infected, and Recovered) at each time ¹.

`plt.plot(t, I)` creates a plot with time `t` on the x-axis and the number of infected individuals `I` on the y-axis. This plot shows how the number of infected individuals changes over time during the simulation.

```python
t, S, I, R = EoN.basic_discrete_SIR(G, 0.6, tmax = 20)
plt.plot(t, S)
```
